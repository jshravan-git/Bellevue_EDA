**Exercise 11-3**

If the quantity you want to predict is a count, you can use Poisson regression, which is implemented in StatsModels with a function called poisson. It works the same way as ols and logit. As an exercise, let’s use it to predict how many children a woman has born; in the NSFG dataset, this variable is called numbabes.

Suppose you meet a woman who is 35 years old, black, and a college graduate whose annual household income exceeds $75,000. How many children would you predict she has born?

From the Excercise question following are identified<br>
**response variable** <br>
    number of childern  -  use attribute  'numbabes' as response variable

**Independent variables or Predictor**<br>
    Age          : 35 years -   use attribute 'ager' - age of respondent<br>
    Education    : college graduate - use attribute 'educat' - Education of respondent<br>
    race         : black - use attribute 'race' <br>
    annual income: $75000  - use 'totincr'  - Totsal income of respondent's family<br>

```python
# Load the Respondent dta
resp_df = nsfg.ReadFemResp()
```

```python
resp_df.shape
```

```python
preg_df.shape
```

```python
# Since the numbabes,  totincr are in respondent dataframe, join both dataframe using caseid as the join condition
Preglive_resp_df = live_df.join(resp_df,  on='caseid', rsuffix='_c')
```

```python
Preglive_resp_df.shape
```

```python
# Check the numbabes feature values 
Preglive_resp_df.numbabes
```

```python
Preglive_resp_df.numbabes.isnull().sum()
```

```python
# fill numbabes with median value
import numpy as np
numbabe_med = Preglive_resp_df.numbabes.median()
Preglive_resp_df['numbabes'] = Preglive_resp_df['numbabes'].replace(np.nan, numbabe_med)
```

```python
#Preglive_resp_df = Preglive_resp_df.dropna(subset=['age_r','educat','race','totincr'])
```

```python
Preglive_resp_df.shape
```

```python
# Build the regression formula with the independent variables and dependent variable
pos_formula = 'numbabes ~ age_r + C(race) + totincr + educat'
# Create the posson distribution model as prediction is needed to find number of babies from the existing data
pos_model = smf.poisson(pos_formula, data=Preglive_resp_df)
#  Fit the variable data and retrieve the results
pos_results = pos_model.fit()
# Print the summary to find which variable has statisticaly significant impact on dependent variable
pos_results.summary()
```

```python
import pandas as pd
# predicting number of babies for woman who is 35 years old, black, and a college graduate whose annual household income exceeds $75,000
pred_columns = ['age_r', 'race', 'totincr', 'educat']
# total income - totincr  value is 14 for $75000 and above
# educat -  4 years college education value is 16
# race -  Black has value 1
pred_df = pd.DataFrame([[35, 1, 14, 16]], columns=pred_columns)
pos_results.predict(pred_df)
```

**Number of babies predicted with given dependent variables in 1.12 babies , can be rounded to 1 baby**

```python
#  Add square of age of respondent to have a better prediction for non-linear relationship with numbabes
Preglive_resp_df['age_sq'] = Preglive_resp_df.age_r**2
```

```python
# Build the regression formula with the independent variables including a derived age square variable and dependent variable
pos_formula_new = 'numbabes ~ age_r + age_sq + C(race) + totincr + educat'
pos_model_new = smf.poisson(pos_formula_new, data=Preglive_resp_df)
pos_results_new = pos_model_new.fit()
pos_results_new.summary()
```

```python
# predicting number of babies for woman who is 35 years old, black, and a college graduate whose annual household income 
# exceeds $75,000 and including square of ages derived value for better non-linear prediction
pred_columns = ['age_r', 'age_sq', 'race', 'totincr', 'educat']
# total income - totincr  value is 14 for $75000 and above
# educat -  4 years college education value is 16
# race -  Black has value 1
# age is 35
# age_sq value is square of 35
pred_df = pd.DataFrame([[35, 35**2, 1, 14, 16]], columns=pred_columns)
pos_results_new.predict(pred_df)
```

**Number of babies predicted with given dependent variables in 1.47 babies or 1.5 babies (Woman can have 1 or 2 babies)**

**Exercise 11-4**

If the quantity you want to predict is categorical, you can use multinomial logistic regression, which is implemented in StatsModels with a function called mnlogit. As an exercise, let’s use it to guess whether a woman is married, cohabitating, widowed, divorced, separated, or never married; in the NSFG dataset, marital status is encoded in a variable called rmarital.

Suppose you meet a woman who is 25 years old, white, and a high school graduate whose annual household income is about $45,000. What is the probability that she is married, cohabitating, etc?

From the Excercise question following are identified<br>
**response variable** is martial status -  attribute is 'rmartial'

**Independent variables**<br>
    Age          : 25 years -   use attribute 'age_r' - age of respondent<br>
    Education    : high school graduate - use attribute 'educat' - Education of respondent<br>
    race         : white - use attribute 'race' <br>
    annual income: $45000  - use 'totincr'  - Totsal income of respondent's family<br>


```python
# Build the regression formula with the independent variables and dependent variable
logit_formula='rmarital ~ age_r  + C(race) + totincr + educat'
# Create a multinomial logit regrssion  model as prediction is needed to find marital status from the existing data
# Since a categorical non-binary classification is required to be predicted multinomial logit regrssion is used 
logit_model = smf.mnlogit(logit_formula, data=Preglive_resp_df)
#  Fit the variable data and retrieve the results
logit_results = logit_model.fit()
# Print summary to find the significance impact of independent variables for each categorical classification response variable values
logit_results.summary() 
```

```python
# Build a input datframe with the given inputs 
pred_logit_columns = ['age_r',  'race', 'totincr', 'educat']
# total income - totincr  value is 11 for $45000 income
# educat -  high school diploma education value is 12
# race -  White has value 1
# age is 25
pred_logit_df = pd.DataFrame([[25, 2, 11, 12]], columns=pred_logit_columns)
logit_results.predict(pred_logit_df)
```

**The prediction is**<br>
The woman's marital status is  <br>
- 'Currently Married' with 63% probability<br>
- 'Not Married but living with opp sex partner' with 10% probability<br>
- 'Divorced' with 11% probability<br>
- 'Never been Married' with 8% probability<br>

```python
# Build the regression formula with the independent variables, including squared values of age to aid non-linear relationship
# predictions and dependent variable
logit_formula='rmarital ~ age_r + age_sq + C(race) + totincr + educat'
# Create a multinomial logit regrssion  model
logit_model = smf.mnlogit(logit_formula, data=Preglive_resp_df)
#  Fit the variable data and retrieve the results
logit_results = logit_model.fit()
# Print summary to find the significance impact of independent variables for each categorical classification response variable values
logit_results.summary() 
```

```python
# predicting marital status of a woman who is 25 years old, white, and a school diploma graduate whose annual household income is $45,000
pred_logit_columns = ['age_r', 'age_sq', 'race', 'totincr', 'educat']
# total income - totincr  value is 11 for $45000 income
# educat -  high school diploma education value is 12
# race -  White has value 1
# age is 25
# age_sq value is square of 25
pred_logit_df = pd.DataFrame([[25, 25**2, 2, 11, 12]], columns=pred_logit_columns)
logit_results.predict(pred_logit_df)
```

**No change noticed in predicted probabilites ,  even if the age value is squared and added as another independent variable for the response variable**

**The prediction is**<br>
The woman's marital status is  <br>
- 'Currently Married' with 63% probability<br>
- 'Not Married but living with opp sex partner' with 10% probability<br>
- 'Divorced' with 11% probability<br>
- 'Never been Married' with 9% probability<br>
